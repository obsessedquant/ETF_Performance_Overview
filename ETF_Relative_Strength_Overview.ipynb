```python
# nasdaq_data = yf.download('^IXIC',period='2y',interval='1d')
# data = yf.download('^GSPC',period='2y',interval='1d')
nasdaq_data = yf.download('^IXIC',period='2y',interval='1wk')
data = yf.download('^GSPC',period='2y',interval='1wk')

df = pd.merge(nasdaq_data,data,on='Date',suffixes=('_NASDAQ','_SP500'))
df = df.drop(df.columns[[0,1,2,4,5,6,7,8,10,11]],axis=1, inplace=False)
df['rs'] = df['Close_SP500'] / df['Close_NASDAQ']
df['rs_ma'] = df['rs'].astype(float).rolling(50).mean()
df['SP500'] = (df['rs'] - df['rs_ma']) / df['rs_ma'] * 100
rs_df = df.drop(df.columns[[0,1,2,3]],axis=1, inplace=False)

for ticker in tickers:
    print(ticker)
#     data = yf.download(ticker,period='max',interval='1d')
    data = yf.download(ticker,period='max',interval='1wk')
    df = pd.merge(nasdaq_data,data,on='Date',suffixes=('_NASDAQ','_'+ticker))
    df = df.drop(df.columns[[0,1,2,4,5,6,7,8,10,11]],axis=1, inplace=False)
    df['rs'] = df['Close_'+ticker] / df['Close_NASDAQ']
    df['rs_ma'] = df['rs'].astype(float).rolling(50).mean()
    df[ticker] = (df['rs'] - df['rs_ma']) / df['rs_ma'] * 100
    df = df.drop(df.columns[[0,1,2,3]],axis=1, inplace=False)
    rs_df = pd.merge(rs_df,df,on='Date',how='outer')

# rs_df = rs_df[550:]
rs_df.to_csv('Output/ETF_rs.csv')
```

```python
# nasdaq_data = yf.download('^IXIC',period='2y',interval='1d')
# data = yf.download('^GSPC',period='2y',interval='1d')
nasdaq_data = yf.download('^IXIC',period='max',interval='1d')
data = yf.download('^GSPC',period='max',interval='1d')

df = pd.merge(nasdaq_data,data,on='Date',suffixes=('_NASDAQ','_SP500'))
df.to_csv('test1.csv')
df = df.drop(df.columns[[0,1,2,4,5,6,7,8,10,11]],axis=1, inplace=False)
df['rs'] = df['Close_SP500'] / df['Close_NASDAQ']
df['rs_20ma'] = df['rs'].astype(float).rolling(20).mean()
df['rs_50ma'] = df['rs'].astype(float).rolling(50).mean()
df['SP500'] = (df['rs'] - df['rs_50ma']) / df['rs_50ma'] * 100
df['SP500_Z'] = (df['SP500'] - df['SP500'].mean()) / df['SP500'].std()
df.to_csv('test2.csv')
rs_df = df
# rs_df = df.drop(df.columns[[0,1,2,3]],axis=1, inplace=False)
rs_df['SP500_20d_v'] = ''
rs_df['SP500_20d_a'] = ''
rs_df['SP500_20d_j'] = ''
rs_df['SP500_20d_vpct'] = ''
rs_df['SP500_20d_apct'] = ''
rs_df['SP500_20d_jpct'] = ''
rs_df['SP500_20v'] = ''
rs_df['SP500_20a'] = ''
rs_df['SP500_20j'] = ''
rs_df['SP500_50d_v'] = ''
rs_df['SP500_50d_a'] = ''
rs_df['SP500_50d_j'] = ''
rs_df['SP500_50d_vpct'] = ''
rs_df['SP500_50d_apct'] = ''
rs_df['SP500_50d_jpct'] = ''
rs_df['SP500_50v'] = ''
rs_df['SP500_50a'] = ''
rs_df['SP500_50j'] = ''

# 20 day
# Velocity, Acceleration, Jerk calculations
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_20d_v'][x] = (rs_df['rs_20ma'][x] - rs_df['rs_20ma'][x-1])
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_20d_a'][x] = (rs_df['SP500_20d_v'][x] - rs_df['SP500_20d_v'][x-1])
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_20d_j'][x] = (rs_df['SP500_20d_a'][x] - rs_df['SP500_20d_a'][x-1])
    
# Velocity, Acceleration, Jerk pct calculations
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_20d_vpct'][x] = (rs_df['SP500_20d_v'][x]) / rs_df['rs_20ma'][x-1]
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_20d_apct'][x] = (rs_df['SP500_20d_a'][x]) / rs_df['SP500_20d_v'][x-1]
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_20d_jpct'][x] = (rs_df['SP500_20d_j'][x]) / rs_df['SP500_20d_a'][x-1]

# Velocity, Acceleration, Jerk Z-Score calculations
rs_df['SP500_20v'] = (rs_df['SP500_20d_vpct'] - rs_df['SP500_20d_vpct'].mean()) / rs_df['SP500_20d_vpct'].std()
rs_df['SP500_20a'] = (rs_df['SP500_20d_apct'] - rs_df['SP500_20d_apct'].mean()) / rs_df['SP500_20d_apct'].std()
rs_df['SP500_20j'] = (rs_df['SP500_20d_jpct'] - rs_df['SP500_20d_jpct'].mean()) / rs_df['SP500_20d_jpct'].std()


# 50 day
# Velocity, Acceleration, Jerk calculations
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_50d_v'][x] = (rs_df['rs_50ma'][x] - rs_df['rs_50ma'][x-1])
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_50d_a'][x] = (rs_df['SP500_50d_v'][x] - rs_df['SP500_50d_v'][x-1])
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_50d_j'][x] = (rs_df['SP500_50d_a'][x] - rs_df['SP500_50d_a'][x-1])
    
# Velocity, Acceleration, Jerk pct calculations
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_50d_vpct'][x] = (rs_df['SP500_50d_v'][x]) / rs_df['rs_50ma'][x-1]
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_50d_apct'][x] = (rs_df['SP500_50d_a'][x]) / rs_df['SP500_50d_v'][x-1]
for x in range(len(rs_df['SP500'])):
    rs_df['SP500_50d_jpct'][x] = (rs_df['SP500_50d_j'][x]) / rs_df['SP500_50d_a'][x-1]

# Velocity, Acceleration, Jerk Z-Score calculations
rs_df['SP500_50v'] = (rs_df['SP500_50d_vpct'] - rs_df['SP500_50d_vpct'].mean()) / rs_df['SP500_50d_vpct'].std()
rs_df['SP500_50a'] = (rs_df['SP500_50d_apct'] - rs_df['SP500_50d_apct'].mean()) / rs_df['SP500_50d_apct'].std()
rs_df['SP500_50j'] = (rs_df['SP500_50d_jpct'] - rs_df['SP500_50d_jpct'].mean()) / rs_df['SP500_50d_jpct'].std()

rs_df2 = rs_df.drop(rs_df.columns[[0,1,2,3,4,5,7,8,9,10,11,12,16,17,18,19,20,21]],axis=1, inplace=False)

rs_df2.to_csv('test.csv')
rs_df2.tail()
```
